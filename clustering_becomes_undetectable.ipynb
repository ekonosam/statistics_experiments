## Clustering becomes undetectable

A sample is a collection of random variables $X_{1}, X_{2}, \dots, X_{n}$ They **do not** necessarily need to have the same distribution. 

For example, when you have a dataset, each row represents realizations of random variables, and some of them may follow different distributions across different groups. For instance, the salaries of men and women may come from distinct distributions.

```python
from scipy.stats import multivariate_normal
import numpy as np
import pandas as pd
# Salaries (latent variable)
men_sal= multivariate_normal([4,4] ,[[1,0],[0,1]])
wom_sal =multivariate_normal([1,5] ,[[1,0],[0,1]])
# Generate full dataset
dataset= np.concatenate((men_sal.rvs(size=100),wom_sal.rvs(size=100)),axis=0)
dataframe= pd.DataFrame(dataset, columns=["Salaries","Seniority"])

# Visualizar dataset
import seaborn as sns
sns.scatterplot(x=dataframe["Salaries"],y=dataframe["Seniority"])
```

The dataset contains 200 bivariate samples. The first 100 are drawn from a normal distribution with mean ([4, 4]), and the next 100 from one with mean ([1, 5]); both use the covariance matrix [[1,0],[0,1]]
